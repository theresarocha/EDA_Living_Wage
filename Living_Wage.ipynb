```python
for i in list:
  def cities_lat(i):
    return geolocator.geocode(i).latitude

data['latitude'] = data['city'].apply(cities_lat)
```

```python
for i in list:
  def cities_long(i):
    return geolocator.geocode(i).longitude

data['longitude'] = data['city'].apply(cities_long)
```

```python
top_cities = data.groupby('city').sum().sort_values(by=['population_2020'], ascending = False)
```

```python
import plotly.express as px
import plotly.io as pio
```

```python
#Customizing chart using mapbox scatter plot
fig = px.scatter_mapbox(data, lat = 'latitude', lon = 'longitude', hover_name = 'city', 
                        hover_data = ['population_2020', 'state'], color_discrete_sequence = ['green'],
                        center = {'lat': 38.2617, 'lon': -96.7073}, size = 'population_2020', 
                        zoom = 3.1, height = 400, width = 1000, size_max = 50, opacity = 0.7)

#Customizing style and margin
fig.update_layout(mapbox_style = 'open-street-map')
fig.update_layout(margin = {'r': 0, 't': 0, 'l': 0, 'b': 0})

#Chart plot
pio.show(fig, renderer = 'jupyterlab')
```
