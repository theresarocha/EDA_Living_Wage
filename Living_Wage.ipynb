#**Living wage of the largest cities in the United States**

Exploratory data analysis with some information about the population of the 100 largest cities in the United States.

This analysis has the intention to answer the following questions:

- Which city has the largest population according to the 2020 United States Census?

- Which city has the smallest number of residents?

- Which state has the lowest population density? And which has the largest number of residents per square mile?

- Which city and state have the largest land area per square mile? And which ones have the smallest area?

- What is the average hourly minimum wage per adult who doesn't have children?

- What is the average hourly minimum wage per adult who has one or more children?



```python
#Importing dataset

import pandas as pd

url = 'https://raw.githubusercontent.com/theresarocha/EDA_Living_Wage/main/dataset/livingwage%20(2).csv'
data = pd.read_csv(url)
```

```python
#Visualizing the first 10 rows in the dataset

data.head(10)
```

```python
#Amount of rows and columns

data.shape
```

```python
#Information about data variables

data.info()
```

The dataset has 19 variables, which are integer, float and object.

The variables are the following ones:

- Rank_2020 = City rank by population as of April 1, 2020, as enumerated by the United States Census Bureau.
- City = City name.
- State = State name of each city.
- Population_2020 = City population as of April 1, 2020, as enumerated by the 2020 United States Census.
- Population_2010 = City population as of April 1, 2020, as enumerated by the 2010 United States Census.
- Land_area_sqmi = City land area as of January 1, 2020.
- Density = City population density as of April 1, 2020 (residents per unit of land area).
- One_adult_no_kids_living_wage = Living wage for one adult and no kids.
- One_adult_one_kid_living_wage = Living wage for one adult and one kid.
- One_adult_two_kids_living_wage = Living wage for one adult and two kids.
- One_adult_three_kid_living_wage = Living wage for one adult and three kids.
- Two_adults_one_working_no_kids_living_wage = Living wage for two adults, one working, and no kids.
- Two_adults_one_working_one_kid_living_wage = Living wage for two adults, one working, and one kid.
- Two_adults_one_working_two_kids_living_wage = Living wage for two adults, one working, and two kids.
- Two_adults_one_working_three_kids_living_wage = Living wage for two adults, one working, and three kids.
- Two_adults_both_working_no_kids_living_wage = Living wage for two adults, both working, and no kids.
- Two_adults_both_working_one_kid_living_wage = Living wage for two adults, both working, and one kid.
- Two_adults_both_working_two_kids_living_wage = Living wage for two adults, both working, and two kids.
- Two_adults_both_working_three_kids_living_wage = Living wage for two adults, both working, and three kids.

```python
#Importing geopy library to get coordinates

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="myapplication")

#Making a list of the 100 largest cities in US

list = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix',
       'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose',
       'Austin', 'Jacksonville', 'Fort Worth', 'Columbus', 'Indianapolis',
       'Charlotte', 'San Francisco', 'Seattle', 'Denver', 'Washington',
       'Nashville', 'Oklahoma City', 'El Paso', 'Boston', 'Portland',
       'Las Vegas', 'Detroit', 'Memphis', 'Louisville', 'Baltimore',
       'Milwaukee', 'Albuquerque', 'Tucson', 'Fresno', 'Sacramento',
       'Kansas City', 'Mesa', 'Atlanta', 'Omaha', 'Colorado Springs',
       'Raleigh', 'Long Beach', 'Virginia Beach', 'Miami', 'Oakland',
       'Minneapolis', 'Tulsa', 'Bakersfield', 'Wichita', 'Arlington',
       'Aurora', 'Tampa', 'New Orleans', 'Cleveland', 'Honolulu',
       'Anaheim', 'Lexington', 'Stockton', 'Corpus Christi', 'Henderson',
       'Riverside', 'Newark', 'Saint Paul', 'Santa Ana', 'Cincinnati',
       'Irvine', 'Orlando', 'Pittsburgh', 'St. Louis', 'Greensboro',
       'Jersey City', 'Anchorage', 'Lincoln', 'Plano', 'Durham',
       'Buffalo', 'Chandler', 'Chula Vista', 'Toledo', 'Madison',
       'Gilbert', 'Reno', 'Fort Wayne', 'North Las Vegas',
       'St. Petersburg', 'Lubbock', 'Irving', 'Laredo', 'Winston–Salem',
       'Chesapeake', 'Glendale', 'Garland', 'Scottsdale', 'Norfolk',
       'Boise', 'Fremont', 'Spokane', 'Santa Clarita', 'Baton Rouge',
       'Richmond']
```

```python
#Getting latitude of all these cities

for i in list:
  def cities_lat(i):
    return geolocator.geocode(i).latitude

data['latitude'] = data['city'].apply(cities_lat)
```

```python
#Getting longitude of all cities

for i in list:
  def cities_long(i):
    return geolocator.geocode(i).longitude

data['longitude'] = data['city'].apply(cities_long)
```

##**Which city has the largest population according to the 2020 United States Census?**


```python
#Sum of population in 2020 by city

top_cities = data.groupby('city').sum().sort_values(by=['population_2020'], ascending = False)
```

```python
#Visualizing the 10 lagest cities considering population in 2020

top_cities.head(10)
```

```python
#Importing seaborn and matplotlib libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Ranking the 10 largest cities considering population in 2020
top10_cities = data.sort_values(by = ['population_2020'], ascending = False)['city'][:10]
barplot = sns.barplot(x = top10_cities, y = data['population_2020'], data = top10_cities, ci = None,  estimator = np.sum, edgecolor = 'gray', color = 'green')

#Chart and labels' title
plt.title('The 10 Largest Cities', fontsize = 15)
plt.ylabel('Population in 2020', fontsize = 13)
plt.xlabel('Cities', fontsize = 13)

#Chart size
plt.gcf().set_size_inches(13,6)

#Labels
plt.bar_label(barplot.containers[0])
plt.ticklabel_format(axis = 'y', style = 'plain')

#Chart plot
plt.show()
```
