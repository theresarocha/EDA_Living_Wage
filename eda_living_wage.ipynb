```python
#Getting latitude of all these cities

from geopy.exc import GeocoderTimedOut
geolocator = Nominatim(user_agent="myapplication", timeout=10)

def get_latitude(city):
    try:
        location = geolocator.geocode(city)
        return location.latitude if location else None
    except GeocoderTimedOut:
        return get_latitude(city)

data['latitude'] = data['city'].apply(get_latitude)
```

```python
#Getting longitude of all cities

def get_longitude(city):
    try:
        location = geolocator.geocode(city)
        return location.longitude if location else None
    except GeocoderTimedOut:
        return get_longitude(city)

data['longitude'] = data['city'].apply(get_longitude)
```

##**What is the average hourly minimum wage per one adult?**



```python
one_adult_concat = data.melt(value_vars =
                             ['one_adult_no_kids_living_wage',
                              'one_adult_one_kid_living_wage',
                              'one_adult_two_kids_living_wage',
                              'one_adult_three_kids_living_wage'],
                             var_name = 'adult', value_name = 'living_wage',
                             ignore_index = True)
```

```python
#Creating the violin plot
fig = px.violin(one_adult_concat, y='adult', x='living_wage', box=True,
                points='all',
                labels={'living_wage': 'Average Living Wage',
                        'adult': 'Adult Type'},
                title='Average Living Wage by One Adult')

#Chart plot
fig.show()
```

Here is a violin plot visualization using a categorical feature (Adult Type) and a numeric feature (Average Living Wage).

- Considering one adult that has 3 kids, we can see that the greater data concentration is in between 40 to 50 dollars/h. The median is 49.125. In other words, 50% of adults with 3 kids has an average living wage up to 49.125 dollars per hour.
Quartile 1 (Q1) is 46.565. 25% of them has an average living wage up to 46.565 dollars/h. And Quartile 3 (Q3) is 57.055. 75%  of them has an average living wage up to 57 dollars/h. The maximum is 81.63 dollars/h and minimun is 40.57 dollars/h.


- Considering one adult that has 2 kids, we can see that the greater data concentration is in between 30 to 40. 50% of one adult with 2 kids has an average living wage up to 38.06 dollars/h.
25% of them has an average living wage up to 35.85 dollars/h. And 75%  of them has an average living wage up to 43.18 dollars/h. The maximum is 61.7 dollars/h and minimun is 31.7 dollars/h.


- When analizing one adult with just 1 kid, median is 30.66 dollars/h, Q1 is 29.19 dollars/h and Q3 is 33.95 dollars/h. The greater concentration of data is between 27 to 34 dollars/h. The maximum living wage is 50.18 dollars/h and minimum is 26.18 dollars/h.

- Finally, analizing one adult that doesn't have kids, we can see the greater data concentration is between 10 to 20 dollars/h. This plot, specifically, is greater in terms of area comparing to the other ones. The standard deviation is smaller then the others and, as a result, the lower and upper limits are smaller comparing to the other graphs.
50% of adults that doesn't have kids has an average living wage up to 15 dollars/h. 25% of them has an average living wage up to 13.93 dollars/h and 75% has an average living wage up to 16.33 dollars/h.
Minimum average living wage is 12.4 and maximum is 25.12.

##**Conclusion**

From the analyzes carried out, New York City boasts the highest population in the United States, with 8,804,190 million inhabitants, while Richmond is the least populated, with 230 thousand inhabitants. Additionally, New York City has the highest population density, with 29,298 residents per unit of land area. In contrast, Anchorage in Alaska, has the lowest population density, with only 171 residents per unit of land area.


About the average living wage in the US, the analysis of the Census 2020 data reveals an interesting trend: as the number of children in a family group increases, so does the average family income. For example, among families with one adult and three children, 50% of them earn an average living wage of up to 49.125 dollars per hour. In contrast, for one adult without children, the average living wage for 50% of them is up to 15 dollars per hour. This correlation persists even among families with two adults, where both are employed or only one adult works.
